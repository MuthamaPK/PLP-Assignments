```python
import pandas as pd

data = pd.read_csv('Book1.csv')
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
grouped = df.groupby('City').agg({'Marks': 'mean', 'Name': 'count'})
```

```python
import matplotlib.pyplot as plt

#plotting name vs marks
plt.plot(df['Name'], df['Marks'])
plt.xlabel('Name')
plt.ylabel('Marks')
plt.title('Name vs Marks')
plt.show()
```

```python
df.groupby('City')['Marks'].mean().plot(kind='bar')
plt.xlabel('Marks')
plt.ylabel('Average Marks')
plt.title('Average Marks vs City')
```

```python
df.plot(kind='scatter', x='Name', y='Marks')
plt.title('Name vs Marks')
```
